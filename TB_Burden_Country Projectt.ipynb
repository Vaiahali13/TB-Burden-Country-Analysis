### 1)Distribution of Prevalence of TB per 100k

```python
# Distribution analysis
plt.figure(figsize=(10, 6))
sns.histplot(df['prevalence_of_TB_100k'], bins=20, kde=True)
plt.title('Distribution of Prevalence of TB per 100k')
plt.xlabel('Prevalence of TB per 100k')
plt.ylabel('Frequency')
plt.show()
```

### 2)Relationship between Total Population and Deaths from TB

```python
correlation = df['total_population'].corr(df['deaths_from_TB'])
print("Correlation Coefficient between Total Population and Deaths from TB:", correlation)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_population', y='deaths_from_TB', data=df)
plt.title('Relationship between Total Population and Deaths from TB')
plt.xlabel('Total Population')
plt.ylabel('Deaths from TB')
plt.show()
```

### 3)What is the trend in TB prevalence year wise?

```python


# Group by year and calculate the total TB prevalence for each year
prevalence_trend = df.groupby(['Year'])['prevalence_of_TB_100k'].sum()

print(prevalence_trend)
# Plotting the trend graph
plt.figure(figsize=(10, 6))
plt.plot(prevalence_trend.index,prevalence_trend.values, marker='o', color='b')
plt.title(' Total TB prevalence for each year')
plt.xlabel('Year')
plt.ylabel('Total TB Prevalence (per 100,000 population)')
plt.grid(True)
plt.show()
```

```python
df["Region"].unique()
```

### 4)What are the TB prevalence rates across different regions?

```python
grouped_by_region = df.groupby('Region')

# Calculate the average TB prevalence for each region
average_prevalence_by_region = grouped_by_region['prevalence_of_TB_100k'].mean()
print(average_prevalence_by_region)
# Visualize the results (bar plot)
average_prevalence_by_region.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Average TB Prevalence by Region')
plt.xlabel('Region')
plt.ylabel('Average TB Prevalence (per 100,000 population)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show plot
plt.show()
```

 ### 5)Which top 10 countries have the highest TB prevalence rates?

```python
country_max_prevalence = df.groupby('Country')['prevalence_of_TB_100k'].max()

# Sort the countries based on their maximum prevalence rate and select the top 10
top_10_countries = country_max_prevalence.sort_values(ascending=False).head(10)
print(top_10_countries )
# Plotting the bar graph
plt.figure(figsize=(12, 8))
colors = ['green', 'lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue','brown','blue','skyblue','red']
plt.bar(top_10_countries.index, top_10_countries.values, color=colors)
plt.title('Top 10 Countries with Highest TB Prevalence Rates')
plt.xlabel('Country')
plt.ylabel('TB Prevalence Rate (per 100,000 population)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show plot
plt.show()

```

### 8)Which country has the highest mortality rate from TB?

### 9)What is the average mortality rate of TB per 100k population?

```python
# Calculate average TB mortality rate
average_mortality_rate = df['mortality_of_TB_100k'].mean()
print("Average mortality rate of TB per 100k population:", average_mortality_rate)
# Plotting histogram of TB mortality rates
plt.figure(figsize=(10, 6))
plt.hist(df['mortality_of_TB_100k'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of TB Mortality Rates')
plt.xlabel('TB Mortality Rate per 100k')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

### 10)Is there any trend in TB mortality rates over time?

```python
# Plotting trend of TB mortality rates over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='mortality_of_TB_100k', data=df)
plt.title('Trend of TB Mortality Rates Over Time')
plt.xlabel('Year')
plt.ylabel('TB Mortality Rate per 100k')
plt.show()
```

### 11)correlation between TB incidence and mortality?

```python
correlation_tb = df[['incidence_per_100k', 'mortality_of_TB_100k']].corr().iloc[0, 1]
print("Correlation between TB incidence and mortality:", correlation_tb)
plt.figure(figsize=(10, 6))
plt.scatter(df['incidence_per_100k'], df['mortality_of_TB_100k'], color='skyblue', alpha=0.7)
plt.title('TB Incidence vs. Mortality Rate')
plt.xlabel('TB Incidence per 100k')
plt.ylabel('TB Mortality Rate per 100k')
plt.grid(True)
plt.show()
```

### 12)How does the TB mortality rate differ across different regions?

### 16)How do the reported low and high bounds of deaths from TB compare to the actual reported values?


```python
# Calculate the difference between actual and low bound deaths from TB
difference_low = df['deaths_from_TB'] - df['deaths_from_TB_low']
# Calculate the difference between actual and high bound deaths from TB
difference_high = df['deaths_from_TB_high'] - df['deaths_from_TB']

# Plotting the differences
plt.figure(figsize=(10, 6))
plt.scatter(difference_low, difference_high, color='skyblue')
plt.title('Comparison of Low and High Bounds of Deaths from TB')
plt.xlabel('Difference from Low Bound')
plt.ylabel('Difference from High Bound')
plt.grid(True)
plt.show()
```

### 17)correlation between mortality rate due to HIV and TB mortality rates?

### 18)How do the reported case detection rates of TB compare to the actual reported values?

```python
# Calculate the difference between reported case detection rates and actual reported values
difference_detection_actual = df['Case_detect_rate_percent'] - df['Case_detect_rate_percent_low']

# Plotting the differences
plt.figure(figsize=(10, 6))
plt.hist(difference_detection_actual, bins=20, color='skyblue', edgecolor='black')
plt.title('Comparison of Reported Case Detection Rates and Actual Reported Values')
plt.xlabel('Difference in Reported Case Detection Rates')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
